# Derivatives

A derivative measures how much a function output changes when its input changes. In machine learning, derivatives tell us how to change parameters to reduce a loss function.

## Definition

For a scalar function `f(x)`, the derivative at `x` is:

$$ f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} $$

The fraction is the slope between two nearby points. The limit asks what happens as the two points become extremely close.

```python
import numpy as np


def f(x: np.ndarray) -> np.ndarray:
    return x**2


def numerical_derivative(x: np.ndarray, h: float = 1e-5) -> np.ndarray:
    return (f(x + h) - f(x)) / h


x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])
numerical_derivative(x)
```

For `f(x) = x^2`, the analytical derivative is:

$$ f'(x) = 2x $$

```python
analytical = 2 * x
approximate = numerical_derivative(x)

np.column_stack([x, analytical, approximate])
```

## PyTorch Autograd

PyTorch can track operations and compute derivatives automatically.

```python
import torch

x_torch = torch.tensor(3.0, requires_grad=True)
y = x_torch**2
y.backward()

x_torch.grad
```
